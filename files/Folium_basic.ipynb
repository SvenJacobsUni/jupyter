# JupyterLite (Python GPS Visualisation)
- GPS-Daten wurden mit meiner Bangle.js 2 Smartwatch getrackt und nach CSV exportiert (siehe FahrtNachBonn.csv)
- Diese werden mithilfe der Folium-Libary in Python visualisiert. Wir benutzen dafür ein JupyterLite-Notebook, welches Pythoncode direkt im Browser ausführen kann

Multiple Paths

```python
# Installieren der notwendigen Module
import piplite
await piplite.install("pandas")
await piplite.install("folium")

# Importieren der notwendigen Module in das Projekt
import pandas as pd
import folium
from branca.element import Figure


# Mithilfe von Pandas die Daten aus der .csv-Datein einlesen
data1 = pd.read_csv('FahrtNachBonn.csv')
data1.drop('Time', inplace=True, axis=1) # Die Spalte Time entfernen, da diese nicht benötigt wird
data1.drop('Altitude', inplace=True, axis=1) # Die Spalte Altitude entfernen, da diese nicht benötigt wird
data1.dropna(how="all", inplace=True) # Alle leeren Zeilen entfernen (einige, da die Uhr nicht immer GPS-Empfang hat)
coords1 = data1.to_numpy() # aus den Daten ein zweidimensionales Array generieren

data2 = pd.read_csv('WegInsBuero.csv')
data2.drop('Time', inplace=True, axis=1) # Die Spalte Time entfernen, da diese nicht benötigt wird
data2.drop('Altitude', inplace=True, axis=1) # Die Spalte Altitude entfernen, da diese nicht benötigt wird
data2.dropna(how="all", inplace=True) # Alle leeren Zeilen entfernen (einige, da die Uhr nicht immer GPS-Empfang hat)
coords2 = data2.to_numpy() # aus den Daten ein zweidimensionales Array generieren


# Folium so konfigurieren, dass es als Bild ausgegeben wird
fig5=Figure(height=400,width=750) # Höhe und Breite der Ausgabe
m5=folium.Map(location=[coords[0][0], coords[0][1],],tiles='cartodbpositron',zoom_start=14) #Zentrum der Karte ist die Startposition
fig5.add_child(m5)

# GPS-Daten hinzufügen
f1=folium.FeatureGroup("Mein Auto") # Name der definierten Route (wichtig, wenn mehrere genutzt werden)
line_1=folium.vector_layers.PolyLine(coords1,popup='<b>Mein Auto</b>',tooltip='Mein Auto',color='green',weight=10).add_to(f1) # Farbe und weitere Kosmetik
f1.add_to(m5)
f2=folium.FeatureGroup("Ich") # Name der definierten Route (wichtig, wenn mehrere genutzt werden)
line_2=folium.vector_layers.PolyLine(coords2,popup='<b>Ich</b>',tooltip='Ich',color='blue',weight=10).add_to(f2) # Farbe und weitere Kosmetik
f2.add_to(m5)

folium.LayerControl().add_to(m5)

m5 # Ausgabe der Karte
```

```python
# Installieren der notwendigen Module
import piplite
await piplite.install("pandas")
await piplite.install("folium")

# Importieren der notwendigen Module in das Projekt
import pandas as pd
import folium
from branca.element import Figure


# Mithilfe von Pandas die Daten aus der .csv-Datein einlesen
data = pd.read_csv('FahrtNachBonn.csv')

data.drop('Time', inplace=True, axis=1) # Die Spalte Time entfernen, da diese nicht benötigt wird
data.drop('Altitude', inplace=True, axis=1) # Die Spalte Altitude entfernen, da diese nicht benötigt wird
data.dropna(how="all", inplace=True) # Alle leeren Zeilen entfernen (einige, da die Uhr nicht immer GPS-Empfang hat)

coords = data.to_numpy() # aus den Daten ein zweidimensionales Array generieren

# Folium so konfigurieren, dass 
fig5=Figure(height=400,width=750) # Höhe und Breite der Ausgabe
m5=folium.Map(location=[coords[0][0], coords[0][1],],tiles='cartodbpositron',zoom_start=9) #Zentrum der Karte ist die Startposition
fig5.add_child(m5)
f3=folium.FeatureGroup("Mein Auto") # Name der definierten Route (wichtig, wenn mehrere genutzt werden)
line_3=folium.vector_layers.PolyLine(coords,popup='<b>Mein Auto</b>',tooltip='Mein Auto',color='green',weight=10).add_to(f3) # Farbe und weitere Kosmetik
f3.add_to(m5)
folium.LayerControl().add_to(m5)

m5 # Ausgabe der Karte
```
